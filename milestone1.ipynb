```python
import pandas as pd

file_path = "/content/drive/MyDrive/Combined_LCA_Disclosure_Data_FY2020.csv"

use_cols = [
    "CASE_STATUS",
    "VISA_CLASS",
    "EMPLOYER_NAME",
    "JOB_TITLE",
    "WORKSITE_STATE",
    "PREVAILING_WAGE"
]

df = pd.read_csv(file_path, usecols=use_cols)
```

```python
df = df.sample(n=10000, random_state=42)
df.shape
```

```python
df.head()
df.info()
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

num_cols = ["PREVAILING_WAGE"]
cat_cols = [
    "CASE_STATUS",
    "VISA_CLASS",
    "EMPLOYER_NAME",
    "JOB_TITLE",
    "WORKSITE_STATE"
]

df[num_cols] = SimpleImputer(strategy="median").fit_transform(df[num_cols])
df[cat_cols] = SimpleImputer(strategy="most_frequent").fit_transform(df[cat_cols])
```

```python
import numpy as np

df["processing_time_days"] = np.random.randint(60, 180, size=len(df))
```

```python
clean_path = "/content/drive/MyDrive/h1b_cleaned_10000.csv"
df.to_csv(clean_path, index=False)

print(" Milestone-1 completed and dataset saved")
```
